# **5-) Preprocesamiento de datos**

```python
def convert_categorical_values_to_numerical_from_df(spark_df):
    """
    This function creates a Spark DataFrame with all values as numerical
    spark_df: Spark DataFrame
    return the Spark DataFrame with all values as numerical
    """
    # Convert categorical columns to numerical values
    for categorical_column in ["smoker", "sex", "region"]:
        new_column_name = "{column_name}_index".format(column_name = categorical_column)
        indexer = StringIndexer(inputCol=categorical_column, outputCol=new_column_name)
        spark_df = indexer.fit(spark_df).transform(spark_df)
    
    # Remove categorical columns
    columns_to_drop = ["smoker", "sex", "region"]
    spark_df = spark_df.drop(*columns_to_drop)
    
    # Rename new numerical columns
    spark_df = spark_df.withColumnRenamed("smoker_index", "smoker")
    spark_df = spark_df.withColumnRenamed("sex_index", "sex")
    spark_df = spark_df.withColumnRenamed("region_index", "region")

    return spark_df
    
def create_and_clean_spark_dataframe_from_csv(spark_session):
    """
    This function creates a Spark DataFrame from a CSV file
    spark_session: Spark Session
    return the Spark DataFrame
    """
    spark_df = spark_session \
      .read \
      .format("csv") \
      .option("path", "insurance_data.csv") \
      .option("header", True) \
      .schema(StructType([
                  StructField("age", IntegerType()),
                  StructField("sex", StringType()),
                  StructField("bmi", FloatType()),
                  StructField("children", IntegerType()),
                  StructField("smoker", StringType()),
                  StructField("region", StringType()),
                  StructField("charges", FloatType())])) \
      .load()
    return spark_df

def show_numerical_descriptive_statistics(spark_df):
    """
    This function shows descriptive statistics data
    """
    print("\nShow numerical descriptive statistics: \n")
    spark_df.select("age", "sex", "bmi").describe().show()
    spark_df.select("children", "region", "charges").describe().show()

def plot_correlation_scattered_matrix(spark_df):
    """
    This function plots a scattered matrix to show correlations between variables
    """
    numeric_data = spark_df.select(spark_df.columns).toPandas()
    axs = scatter_matrix(numeric_data, figsize=(7, 7), alpha=0.7);
    n = len(numeric_data.columns)
    for i in range(n):
        # Set vertical plot settings
        v = axs[i, 0]
        v.yaxis.label.set_rotation(0)
        v.yaxis.label.set_ha('right')
        v.set_yticks(())
        
        # Set horizontal plot settings
        h = axs[n-1, i]
        h.xaxis.label.set_rotation(90)
        h.set_xticks(())
    plt.show()

def transform_spark_df_to_features_vector_df(spark_df, remove_smoker=False):
    """
    This function transforms a spark df to a features vector df
    """
    assembler = VectorAssembler(
        inputCols=['age', 'sex', 'bmi', 'children', 'region',
                   'charges'] if remove_smoker else \
                  ['age', 'sex', 'bmi', 'children', 'smoker',
                   'region', 'charges'],
        outputCol='features')

    vector_df = assembler.transform(spark_df)
    vector_df = vector_df.select(['features', 'smoker'])
    return vector_df

def plot_correlation_matrix_heatmap(vector_df):
    """
    This function plots a heatmap correlation matrix
    """
    pearson_matrix = Correlation.corr(vector_df, 'features').collect()[0][0]
    sns.heatmap(pearson_matrix.toArray(), annot=True, fmt=".2f", cmap='viridis')

def show_graphs_statistics(spark_df):
    """
    This function shows graph data
    """
    print("\nShow a correlation matrix with numbers and a scattered correlation matrix: \n")
    vector_df = transform_spark_df_to_features_vector_df(spark_df)
    plot_correlation_matrix_heatmap(vector_df)
    plot_correlation_scattered_matrix(spark_df)

def normalize_data(spark_df):
    """
    This function normalize data before training the model
    """
    vector_df = transform_spark_df_to_features_vector_df(spark_df, remove_smoker=True)
    standard_scaler = StandardScaler(inputCol='features', outputCol='scaled_features')
    scale_model = standard_scaler.fit(vector_df)
    scaled_df = scale_model.transform(vector_df)
    
    return scaled_df

def append_column_to_spark_df(spark_df, column_values_df):
    """
    This function appends a column to a Spark DataFrame
    """
    import pyspark.sql.functions as f
    first_df = column_values_df
    first_df = first_df.withColumn('row_index', f.monotonically_increasing_id())
    second_df = spark_df.withColumn('row_index', f.monotonically_increasing_id())
    second_df = second_df.join(first_df, on=["row_index"]).sort("row_index").drop("row_index")
    return second_df

def convert_normalized_df_to_valid_table_for_db(normalized_df):
    """
    This function converts normalize features vector into a valid DF for DB
    """
    normalized_valid_db_format_df = normalized_df.select('scaled_features', 'smoker')
    normalized_valid_db_format_df = normalized_valid_db_format_df.rdd.map(lambda x:[float(y) for y in x['scaled_features']]+[x['smoker']]).toDF(['age', 'sex', 'bmi', 'children', 'region', 'charges', 'smoker'])
    
    return normalized_valid_db_format_df

def write_spark_df_to_db(spark_df, table_name):
    """
    This function writes Spark dataframe to DB
    """
    spark_df \
        .write \
        .format("jdbc") \
        .mode('overwrite') \
        .option("url", POSTGRESQL_URL) \
        .option("user", POSTGRESQL_USER) \
        .option("password", POSTGRESQL_PASSWORD) \
        .option("dbtable", table_name) \
        .save()

def data_preprocessing(spark_session):
    """
    This function completes data preprocessing before training model
    """
    # Loading and cleaning CSV input file data.
    spark_df = create_and_clean_spark_dataframe_from_csv(spark_session)

    # Show cleaned df schema
    print("\nd-) Loading and cleaning CSV input file data.")
    print("\nd.1-) Definition of schema: \n")
    spark_df.printSchema()
    
    print("d.2-) Show that the data has been loaded successfully by selecting a couple of rows: \n")
    spark_df.show()
    
    # Convert all categorical values to numerical values
    spark_df = convert_categorical_values_to_numerical_from_df(spark_df)
    print("d.3-) Show that the data has been converted successfully from categorical to numerical: \n")
    spark_df.show()

    # Show graphs and descriptive statistics prior to training
    print("e-) Show graphs and descriptive statistics prior to training.")
    show_numerical_descriptive_statistics(spark_df)
    show_graphs_statistics(spark_df)
    print("Análisis de las estadisticas descriptivas: ")
    print("- Se observa que la gran mayoria de variables no se encuentran muy correlacionadas a excepcion de smoker y charges.\n")
    
    # Normalizing the data
    print("\nf-) After normalizing the data, the DataFrame looks like this: \n")
    normalized_df = normalize_data(spark_df)
    normalized_df.show()
    
    # Convert to valid format for DB
    print("\ng-) Writing to DB: ")
    print("\ng.1-) Before writing the normalized Dataframe, is necessary to convert it to a valid table for DB: ")
    normalized_valid_db_df = convert_normalized_df_to_valid_table_for_db(normalized_df)
    print("- After convert it to a valid table for DB, the DataFrame looks like this: \n")
    normalized_valid_db_df.show()
    print("- In order to check that this table was written correctly into DB, you can go to PostgreSQL and check the table named tarea3.")
    print("  It should look exactly the same.\n")
    
    # Writing to DB
    write_spark_df_to_db(normalized_valid_db_df, "tarea3")
```

# **6-) Entrenamiento de modelos**

```python
def read_dataset_from_db(spark_session):
    """
    This function reads the clean dataset from the database
    """
    spark_df = spark_session \
               .read \
               .format("jdbc") \
               .option("url", POSTGRESQL_URL) \
               .option("user", POSTGRESQL_USER) \
               .option("password", POSTGRESQL_PASSWORD) \
               .option("dbtable", "tarea3") \
               .load()
    spark_df.show()
    return spark_df

def split_train_test_data(spark_df):
    """
    This function splits train and test data
    """
    train, test = spark_df.randomSplit([0.8, 0.2], seed=12345)
    return train, test

def set_logistic_regression_parameter_grid(logistic_regression):
    """
    This function set parameter grid for Cross Validation
    """
    param_grid = ParamGridBuilder()\
    .addGrid(logistic_regression.maxIter,[10])\
    .build()
    
    return param_grid

def set_random_forest_classifier_parameter_grid(random_forest):
    """
    This function set parameter grid for Cross Validation
    """
    param_grid = ParamGridBuilder()\
    .addGrid(random_forest.maxDepth, [2, 4])\
    .addGrid(random_forest.maxBins, [20])\
    .addGrid(random_forest.numTrees, [5])\
    .build()
    
    return param_grid

def plot_summary_area_under_roc(summary):
    """
    This function shows the areaUnderROC
    """
    roc = summary.roc.toPandas()
    plt.plot(roc['FPR'],roc['TPR'])
    plt.ylabel('False Positive Rate')
    plt.xlabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()

def show_logistic_regression_model_metrics(training_model):
    """
    This function shows model metrics after training
    """
    print("Coefficients: " + str(training_model.coefficients))
    print("Intercept: " + str(training_model.intercept))
    
    # Obtain the receiver-operating characteristic as a dataframe and areaUnderROC.
    print("\nShow receiver-operating characteristic as a dataframe: \n")
    training_model.summary.roc.show()
    
    # Plot areaUnderROC
    print("Show areaUnderROC value and areaUnderROC graph: \n")
    print("areaUnderROC: " + str(training_model.summary.areaUnderROC))
    plot_summary_area_under_roc(training_model.summary)

def building_model_using_logistic_regression(train_data, test_data):
    """
    This function builds a classification model using
    Logistic Regression
    """
    logistic_regression = LogisticRegression(labelCol="smoker",
                                             featuresCol="features",
                                             maxIter=10)
    
    # Set parameter grid
    parameter_grid = set_logistic_regression_parameter_grid(logistic_regression)
    
    # Set evaluator
    evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction",
                                              labelCol="smoker")
    
    # Set Cross Validation
    cross_validation = CrossValidator(estimator=logistic_regression,
                                      estimatorParamMaps=parameter_grid,
                                      evaluator=evaluator, numFolds=3)
    
    # Run Cross Validations
    cross_validation_model = cross_validation.fit(train_data)

    predict_train_data = cross_validation_model.transform(train_data)
    predict_test_data = cross_validation_model.transform(test_data)
    
    # Show model metrics
    print("- After training the model, these are the Logistic Regression Metrics: \n")
    show_logistic_regression_model_metrics(cross_validation_model.bestModel)

    return predict_train_data, predict_test_data

def building_model_using_random_forest(train_data, test_data):
    """
    This function builds a classification model using
    Random Forest
    """
    random_forest = RandomForestClassifier(labelCol="smoker",
                                           featuresCol="features")
    
    # Set parameter grid
    parameter_grid = set_random_forest_classifier_parameter_grid(random_forest)
    
    # Set evaluator
    evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction",
                                              labelCol="smoker")
    
    # Set Cross Validation
    cross_validation = CrossValidator(estimator=random_forest,
                                      estimatorParamMaps=parameter_grid,
                                      evaluator=evaluator, numFolds=3)
    
    # Run Cross Validations
    cross_validation_model = cross_validation.fit(train_data)

    predict_train_data = cross_validation_model.transform(train_data)
    predict_test_data = cross_validation_model.transform(test_data)
    
    # Show model metrics
    print("Note: PySpark doesnt bring any Random Forest Classifier Metrics.\n")
    
    return predict_train_data, predict_test_data

def select_data_for_training_models(spark_session):
    """
    This function selects the data for training two models
    """
    print("\nh-) Read dataset from DB:")
    clean_spark_df = read_dataset_from_db(spark_session)
    
    print("\ni-) Convert features into one single vector:")
    clean_vector_features_df = transform_spark_df_to_features_vector_df(clean_spark_df, remove_smoker=True)
    clean_vector_features_df.show()
    
    # Split train and test data
    train_data, test_data = split_train_test_data(clean_vector_features_df)
    
    return train_data, test_data

def training_model_1(train_data, test_data):
    """
    This function trains a logistic regression model
    """
    print("\nj-) Building a model using Logistic Regression.")
    predict_train_data, predict_test_data = building_model_using_logistic_regression(train_data, test_data)
    return predict_train_data, predict_test_data

def training_model_2(train_data, test_data):
    """
    This function trains a random forest model
    """
    print("\nk-) Building a model using Random Forest Classifier. \n")
    predict_train_data, predict_test_data = building_model_using_random_forest(train_data, test_data)
    return predict_train_data, predict_test_data
```

# **7-) Evaluación de conjunto de validación**

```python
def evaluating_logistic_regression_model(predict_train_data, predict_test_data):
    """
    This function evaluates the logistic regression model
    """
    print("\nMake predictions and evaluates the Logistic Regression Model: \n")
    
    print("\n1. Make predictions with logistic regression model: ")
    predict_test_data.select("smoker", "prediction").show(20)
    
    print("\n2. Evaluates logistic regression model: \n")
    evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="smoker")
    predict_test_data.select("smoker", "rawPrediction", "prediction", "probability").show(15)
    print("The area under ROC for train data is {}".format(evaluator.evaluate(predict_train_data)))
    print("The area under ROC for test data is {}".format(evaluator.evaluate(predict_test_data)))

def evaluating_random_forest_model(predict_train_data, predict_test_data):
    """
    This function evaluates the random forest model
    """
    print("\nMake predictions and evaluates the Random Forest Classifier Model: \n")
    
    print("\n1. Make predictions with Random Forest Classifier Model: ")
    predict_test_data.select("smoker", "prediction").show(20)
    
    print("\n2. Evaluates Random Forest Classifier Model: \n")
    evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="smoker")
    predict_test_data.select("smoker", "rawPrediction", "prediction", "probability").show(15)
    print("The area under ROC for train data is {}".format(evaluator.evaluate(predict_train_data)))
    print("The area under ROC for test data is {}".format(evaluator.evaluate(predict_test_data)))

def save_model_predictions_into_db(predict_test_data, table_name):
    """
    This function saves predictions into DB
    """
    # Transform dataframe to a valid format for DB
    predict_test_data_valid_db_format_df = predict_test_data.select('features', 'smoker', 'prediction')
    predict_test_data_valid_db_format_df = predict_test_data_valid_db_format_df.rdd.map(lambda x:[float(y) for y in x['features']]+[x['smoker']]+[x['prediction']]).toDF(['age', 'sex', 'bmi', 'children', 'region', 'charges', 'smoker', 'prediction'])
    
    # Write to DB
    write_spark_df_to_db(predict_test_data_valid_db_format_df, table_name)

def analysis_of_results(predict_test_data_1, predict_test_data_2):
    """
    This function analyzes the obtained results
    """
    evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="smoker")
    print("The first step to analyze the results is to compare both area under ROC: \n")
    print("The area under ROC for test data using Logistic Regression is {}".format(evaluator.evaluate(predict_test_data_1)))
    print("The area under ROC for test data using Random Forest is {}".format(evaluator.evaluate(predict_test_data_2)))
    print("\nComparing both results, it is observed that the value obtained with Random Forest is greater and therefore the model used with Random Forest is better.\n")
    
```

# **8-) Función main() para ejecutar el programa principal**

```python
def main():
    """
    This function extract data and train a model using PySpark
    """
    print("Step 1: Entry Data Description")
    show_entry_data_description()
    
    print("Step 2: Data Preprocessing")
    spark_session = create_spark_session()
    data_preprocessing(spark_session)
    
    print("Step 3: Training Models")
    train_data, test_data = select_data_for_training_models(spark_session)
    predict_train_data_1, predict_test_data_1 = training_model_1(train_data, test_data)
    predict_train_data_2, predict_test_data_2 = training_model_2(train_data, test_data)
    
    print("Step 4: Evaluating Models")
    evaluating_logistic_regression_model(predict_train_data_1, predict_test_data_1)
    save_model_predictions_into_db(predict_test_data_1, table_name="modelo1")
    
    evaluating_random_forest_model(predict_train_data_2, predict_test_data_2)
    save_model_predictions_into_db(predict_test_data_2, table_name="modelo2")
    
    print("\n- Análisis de resultados: \n")
    analysis_of_results(predict_test_data_1, predict_test_data_2)

# Execute main program
main()
```
